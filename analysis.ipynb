```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

import pickle

item_list = []
# 示例 list
with open('./netease.filtered.txt') as f:
    song_id_info_map = pickle.load(open('./song_id_info_map.pkl', 'rb'))
    for line in f:
        list(map(lambda x: item_list.append(song_id_info_map[int(x)]), line.strip().split(' ')[1].split(',')))


# 1. 统计频次
freq_with_singer = Counter(item_list)
top_freq = dict(freq_with_singer.most_common(100))
top_songs = set(map(lambda x: x[0], top_freq))
freq = Counter(map(lambda x: x[0], filter(lambda x: x[0] in top_songs, item_list)))
with open('./song.singer.count.txt', 'w') as f:
    for (song, singer), counter in sorted(freq_with_singer.items(), key=lambda x: x[1], reverse=True):
        f.write(f'{song}\t{singer}\t{counter}\n')
font_path = "/System/Library/Fonts/STHeiti Light.ttc"

# 2. 生成词云
wordcloud = WordCloud(font_path=font_path, width=1920, height=1080, background_color='white').generate_from_frequencies(freq).to_file('song.freq.jpg')
```
